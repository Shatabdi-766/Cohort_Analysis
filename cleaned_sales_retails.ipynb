```python
import pandas as pd
```

```python
df = pd.read_csv('e:\Interactive Cares\Online+Retail\COHORT_ANALYSIS SQL\Sales_Retail.csv')

df.head()
```

```python
# Remove rows where InvoiceNo starts with "C" (credit notes or canceled invoices)
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]

# Convert InvoiceNo to numeric (ignore errors for non-numeric values)
df["InvoiceNo"] = pd.to_numeric(df["InvoiceNo"], errors="coerce")

# Drop rows where InvoiceNo is NaN (non-numeric values that couldn't be converted)
df = df.dropna(subset=["InvoiceNo"])

# Convert InvoiceNo to Int64
df["InvoiceNo"] = df["InvoiceNo"].astype("Int64")
```

```python
# Change other column types
df["StockCode"] = df["StockCode"].astype(str)  # Assuming 'any' type is treated as string
df["Description"] = df["Description"].astype(str)
df["Quantity"] = df["Quantity"].astype("int64")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["UnitPrice"] = df["UnitPrice"].astype(float)
df["CustomerID"] = df["CustomerID"].astype("Int64")  # Use pandas' nullable integer type
df["Country"] = df["Country"].astype(str)
```

```python
# Convert InvoiceDate to datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

# Extract date part from InvoiceDate
df["InvoiceDate"] = df["InvoiceDate"].dt.date
```

```python
# Filter rows where Customer ID is not null or empty
df = df[df["CustomerID"].notna() & (df["CustomerID"] != "")]

# Display the final DataFrame
df.head()
```

```python
# Save the final DataFrame to a CSV file
output_csv_path = r"e:\Interactive Cares\Online+Retail\COHORT_ANALYSIS SQL\sales_retail_II_cleaned.csv"
df.to_csv(output_csv_path, index=False)

print(f"DataFrame saved to {output_csv_path}")
```
